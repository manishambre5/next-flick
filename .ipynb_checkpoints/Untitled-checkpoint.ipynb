```python
import pandas as pd
import numpy as np
```

```python
# Loading the tmdb dataset
tmdb1 = pd.read_csv("./tmdb/tmdb_5000_credits.csv")
tmdb2 = pd.read_csv("./tmdb/tmdb_5000_movies.csv")
```

```python
tmdb1.head()
#tmdb2.head()
```

```python
# Joining the 2 datasets using the 'id' column
tmdb1.columns = ['id','title','cast','crew']
tmdb2= tmdb2.merge(tmdb1, on='id')
```

```python
# Cleaning Data
```

```python
#tmdb2[tmdb2['title_x'] != tmdb2['title_y']]
tmdb2 = tmdb2.drop('title_y', axis=1)
tmdb2 = tmdb2.rename(columns={'title_x': 'title'})
tmdb2.head()
```

```python
#tmdb2['status'].value_counts()
tmdb2 = tmdb2[~tmdb2['status'].isin(['Rumored', 'Post Production'])]
```

```python
# Filtering out relevant columns
df = tmdb2[['genres','id','keywords','original_language','original_title','overview','popularity','title','vote_average','vote_count']]
```
